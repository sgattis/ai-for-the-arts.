# Enhancing Accessibility and Inclusion in Museums with Large Language Models

## Introduction
The design I am proposing focuses on using large language models (LLMs) to enhance recommendation systems in cultural heritage spaces, specifically by making museums more accessible, engaging, and inclusive for all visitors. Drawing inspiration from the MAGICAL system (Trichopoulos et al., 2023), which uses GPT-4 to respond intelligently to user queries, my creative design will adapt the LLM for museums to cater to diverse audiences, including those with disabilities, cultural differences, and non-traditional learning needs.

The goal of my creative proposal is to present an innovative interaction system where LLMs can augment the visitor experience, fostering both learning, engagement, and inclusion in the process!

More information on the MAGICAL system can be found here: https://doi.org/10.3390/electronics12183829

I will employ a similar training protocol, which helped MAGICAL enhance the museum experience through providing real-time, intelligent, and context-aware assistance. 

The MAGICAL system makes a start on meeting the needs of different groups, but is not specifically tailored to or designed for diverse populations. For example, it can provide information about infrastructure, such as: this museum 'is wheelchair and pushchair accessible.' However, I will use data from diverse populations to train my system to handle specific needs, such as giving neurodivergent visitors the option to be directed to quieter areas of the museum for a more comfortable experience.

### Human and AI Interaction
In my design, system users will primarily be museum visitors. The LLM recommendation system will act as an interactive guide, providing tailored information to meet the needs of diverse user groups.

Multiple human-AI interactions are envisioned:

- Museum Visitors: Visitors can interact with the system through natural language, asking questions about exhibits, history, or specific areas of interest.

- Museum Staff: Staff can consult the system to give the best reccommendations and solutions to visitors who may have specific needs. 

- LLM Interface: The LLM will be fine-tuned with contextual data from the museum’s archives to provide personalised, accurate, and multilingual recommendations. It will adjust its tone and style based on individual preferences.

- Assistive Technology: LLMs will be integrated with assistive technology such as speech-to-text and screen readers, improving accessibility for people with disabilities (e.g., visual or hearing impairments).

### Multi-Modal Interaction
I would like to propose multiple ways for people to interact with the systen to further an emphasis on accessibility for all populations. 

My creative design would offer:
- Text-Based Interaction: Visitors can engage through text, either by use of kiosks or of personal devices (e.g., phone or tablet)

- Voice Interaction: A hands-free interaction, which would be suitable for users with mobility or reading challenges.

- Interactive Design: There is potential for further development with augmented reality, which could provide visual cues that align with the LLM’s responses, offering a fully immersive experience for visitors. This is something that could be developed in the future once the system is in place. 

### Cultural Inclusivity
My LLM will be trained on diverse cultural and linguistic datasets. This allows the system to:

- Cater to a wider, global audience by supporting multiple languages, not just the language of where the museum is located.

- Respond to different cultural contexts, ensuring that all information and exhibits are respectful to various traditions and histories, and focus on displaying and directing visitors towards artifacts from underrepresented groups that have been lost to history. 

## Value of the Design: Learning, Engagement, and Inclusion
### Enhancing Learning
My system will:

- Personalise Learning, so visitors can receive tailored content based on their interests, learning preferences, and previous interactions. For example, a visitor with an interest in queer art will receive additional information on LGBTQ+ artists and their historical context and where to find them within the museum. 

- Offer audio, text, and video resources in different languages and formats, ensuring accessibility for those with different learning needs and preferences.

### Promoting Engagement
My system will:

- Encourage participation, so visitors can engage in dialogue with the system, asking questions and providing feedback in the process, which can, in turn, improve the accuracy of the responses.

- Encourage a more personalised exploration, as it will suggest different pathways through exhibits based on a visitor’s interests, introducing them to art/artifacts they might have overlooked if not for the assistance. 

### Supporting Inclusion
My system will:

- Be accessible for all.

- Encourage an inclusive representation of diverse cultures, identities, and histories, challenging stereotypes and promoting inclusivity among all visitors.

- Highlight the stories of marginalised communities. 

## Implications, Considerations, and Conclusions

My system’s ability to explore diverse voices and cater towards different populations can change the future of accessibility in museums and cultural heritage institutions, rethinking the narratives that they tell and how they tell them. By including a range of viewpoints for a variety of visitors, the system can challenge traditional exhibits and promote an approach that focuses on accssibility and inclusivity. When developing this system, it is important to note technological concerns, such as the idea that heavy reliance on AI could reduce human interaction in museum spaces and reduce the importance of museum staff and curators. However, I believe my system serves as a tool in filling in spaces that staff cannot always be there for.
