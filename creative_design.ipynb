# Enhancing Accessibility in Museum Archives with Large Language Models: A Focus on Inclusion and Diversity

## Introduction
The design I propose focuses on using large language models (LLMs) to enhance recommendation systems in cultural heritage spaces, making museums more accessible, engaging, and inclusive for all visitors. Drawing inspiration from the MAGICAL system (Trichopoulos et al., 2023), which uses GPT-4 to respond intelligently to user queries, my creative design will adapt LLMs for museums to cater to diverse audiences, including those with disabilities, cultural differences, and non-traditional learning needs.

The goal of my creative proposal is to present an innovative interaction system where LLMs augment the visitor experience, fostering both learning, engagement, and inclusion in the process.

More information on the MAGICAL system can be found here: https://doi.org/10.3390/electronics12183829

# Human and AI Interaction
In this design, the system users will be museum visitors. The LLM-powered recommendation system will serve as an interactive guide, providing tailored information and resources to meet the needs of diverse user groups.

The MAGICAL system makes a start on this, but is not specifically tailored to or designed for diverse groups. For example, providing information such as this museum 'is wheelchair and pushchair accessible.'

For my design, multiple human-AI interactions are envisioned:

- Museum Visitors: Visitors can interact with the system via natural language, asking questions about exhibits, history, or specific areas of interest.

- LLM Interface: The LLM (e.g., GPT-4) will be fine-tuned with contextual data from the museum’s archives to provide personalized, multilingual recommendations. It will adjust its tone and style based on individual preferences, enhancing inclusivity for diverse users.

- Assistive Technology: LLMs will be integrated with speech-to-text, screen readers, and other assistive technologies, improving accessibility for people with disabilities (e.g., visual or hearing impairments).

# Multi-Modal Interaction
I would like to propose multiple ways for people to interact with the LLM to further my emphasis on accessibility for all populations. 

My creative design would offer:
- Text-Based Interaction: Visitors can engage via text, either through kiosks or personal devices (e.g., phone or tablet)

- Voice Interaction: A hands-free interaction, which would be suitable for users with mobility or reading challenges.

- Visual and Interactive Design: There is potential for further development with augmented reality, which could provide visual cues that align with the LLM’s responses, offering a fully immersive experience for visitors. This is something that could be developed in the future once the system is in place. 

# Cultural Inclusivity
My LLM will be trained on diverse cultural and linguistic datasets. This allows the system to:

- Cater to a global audience by supporting multiple languages, not just the language of where the museum is located.

- Respond to diverse cultural contexts, ensuring that all information and exhibits are respectful to different traditions and histories, and focus on pushing artifacts from underrepresented groups that have been lost in history. 

## Value of the Design: Learning, Engagement, and Inclusion
# Enhancing Learning
My system will:

- Personalise Learning, so visitors can receive tailored content based on their interests, learning preferences, and previous interactions. For example, a visitor with an interest in queer art will receive additional information on LGBTQ+ artists and their historical context.

- Offer audio, text, and video resources in different languages and formats, ensuring accessibility for learners with different needs and preferences.

# Promoting Engagement
My system will:

- Encourage participation, so visitors can engage in dialogue with the system, asking questions and providing feedback in the process, which can help the design in turn.

- Encourage a personalised exploration, as it will suggest different pathways through exhibits based on a visitor’s interests, introducing them to art/artifacts they might have overlooked. 

# Supporting Inclusion
My system will:

- Be accessible for all. For example, people with disabilities could benefit from LLMs through accessible interfaces and assistive technologies.

- Encourage an inclusive representation of diverse cultures, identities, and histories, challenging stereotypes and promoting empathy among visitors.

- Highlight the stories of marginalised communities. 

# Implications, Considerations, and Conclusions

The system’s ability to highlight marginalised voices and cultural perspectives can change the future of accessibility in museums and cultural heritage institutions, rethinking the narratives that they tell and how they tell them. By including diverse viewpoints, the system can challenge traditional exhibits and promote a cultural dialogue that focuses on accssibility and inclusivity. When developing this system, it is important to note concerns such as the idea that heavy reliance on AI could reduce human interaction in museum spaces, diminishing the importance of traditional museum staff and curators. However, I believe my system serves as a tool in filling in spaces that staff cannot always be there for.
